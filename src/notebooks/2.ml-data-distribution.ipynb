```python
import numpy as np
import matplotlib.pyplot as plt
```

# Uniform Distribution

```python
np.random.seed(71)
random = np.random.uniform(0.0, 5.0, 250)

random
```

```python
plt.hist(random, 5)
plt.show()
```

```python
random = np.random.uniform(0, 5, 100000)
plt.hist(random, 100)
plt.show()
```

# Normal (Gaussian) Distribution

```python
random = np.random.normal(5, 1, 100000) # Parameters are (mean, standard-deviation, number-of-values)
random
```

```python
plt.hist(random, 100)
plt.show()
```

# Scatter Plot

```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(age, speed)
plt.show()
```

# Random Data Distribution on Scatter Plot

```python
x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)

plt.scatter(x, y)
plt.show()
```

We can see that the dots are concentrated around `5` in x-axis and `10` in y-axis as we've given the mean values 5 and, 10 respectively as well as values in y-axis more widely spread than the x-axis given the values of standard deviation given in each random number collection.
